```python
# Define function to get list of files from a folder
def list_files(folder):
    files = []
    for entry in os.listdir(folder):
        if os.path.isfile(os.path.join(folder, entry)):
            files.append(entry)
    return files
```

```python
# For Loop System
for line in lines: # iterate through all cell lines
    TF_files = [file for file in list_files(f"tf_ChIP_data/processed_ChIP") if line in file]
    histone_files = [file for file in list_files("histone_processed_data") if line in file and "consensus_all.bed" in file]
    #print(TF_files)
    #print(histone_files)

    for TF_file in TF_files:
        TF = TF_file.split('_')[1]
        TF_name = f"tf_ChIP_data/processed_ChIP/{TF_file}"
        #print(TF_name)
        TF_bed = BedTool(TF_name)

        for histone_file in histone_files:
            histone = histone_file.split('_')[1]
            histone_name = f"histone_processed_data/{histone_file}"
            #print(histone_name)
            histone_bed = BedTool(histone_name)

            result = TF_bed.intersect(histone_bed)
            result.saveas(f"intersections/{line}_{TF}_{histone}_intersection.bed")


```
